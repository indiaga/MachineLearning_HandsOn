```python
# 라이브러리 로딩
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
boss_this = pd.read_csv('D:\\MyPython\\boss_forecast\\boss_forecast_20200205.csv',header=2, index_col='PJT코드', 
                        thousands=',', decimal ='.', encoding='euc-kr', skipfooter = 2, engine = 'python',
                        dtype = {"매출귀속부서":np.str,"영업대표":str,"PJT명":str,"End_User":str,"매출처":str,"구분":str,"시작일":str,"종료일":str}
                        )
```

```python
for col in boss_this.columns:
     print("\"{}\":{},".format(col, boss_this[col].dtype.name))
```

```python
# 팀분기합계 컬럼 생성
boss_this_sum = boss_this.groupby('매출귀속부서')['매출_1','GP1_1','매출_2','GP1_2','매출_3','GP1_3','매출_4','GP1_4','매출_5','GP1_5','매출_6','GP1_6'].sum()
boss_this_sum['매출_Q1'] = boss_this_sum['매출_1'] + boss_this_sum['매출_2'] + boss_this_sum['매출_3']  
boss_this_sum['GP1_Q1'] = boss_this_sum['GP1_1'] + boss_this_sum['GP1_2'] + boss_this_sum['GP1_3']  
boss_this_sum['매출_Q2'] = boss_this_sum['매출_4'] + boss_this_sum['매출_5'] + boss_this_sum['매출_6']  
boss_this_sum['GP1_Q2'] = boss_this_sum['GP1_4'] + boss_this_sum['GP1_5'] + boss_this_sum['GP1_6']  
boss_this_sum
```
